```python
# Extract column names from the second column to the end and convert it to a list
cols = data.columns[1:].tolist()

# Display the result
print(cols)
```

```python
colors = plt.cm.plasma(np.linspace(0, 1, len(cols)))
for i, color in zip(cols, colors):
  fig=plt.figure(figsize=(10,6))
  plt.hist(data[i], bins=30, color= color, edgecolor='black', alpha=0.7)
  plt.title(f'Distribution of Population in {i}', fontsize=16, color='black', fontweight='bold')
  plt.xlabel(i)
  plt.ylabel('Frequency')
  plt.grid(axis='y')
  plt.show()
```

```python
years = data.columns[1:]
total_population = data[years].sum()
colors = plt.cm.viridis(np.linspace(0, 1, len(years)))
plt.figure(figsize=(30, 30))
plt.bar(years,total_population, color=colors, edgecolor='black')
plt.title('Total Population Over Time', fontsize=24, fontweight='bold', color='darkred')
plt.xlabel('Year', size=20)
plt.ylabel('Total Population', size=20)
plt.xticks(rotation=90, ha='center')
plt.show()
```

```python
# Define the year to visualize
year = '2020'
# Sort countries by population for the given year
population_by_country = data[['Country Name', year]].sort_values(by=year, ascending=False)
# Select the top 10 countries for better clarity
top_countries = population_by_country.head(10)

plt.figure(figsize=(12, 8))
plt.barh(top_countries['Country Name'], top_countries[year], color='lightcoral', edgecolor='black')
plt.title(f'Population by Country in {year}')
plt.xlabel('Population')
plt.ylabel('Country')
plt.grid(axis='x')
plt.show()
```
